## Machine Learning model

```python
car_model.head()
```

```python
car_model['cc'].unique()
```

```python
#step 1 -> Installation of libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , OneHotEncoder

import random 
import numpy as np
```

```python
#car_model - > data is clean and clear for  encoding and scaling

car_model.head()
```

```python
#1. define the category set
category_set = ['fuel_type','cc','doors','cylinders','gears']

#2. create encoder instance
encoder = OneHotEncoder(drop='first' , sparse_output=False) #drop='first' -> it  will drop on extra dummy variable if exist.
                                                     #sparsh=False -> it will return the data in form of numpy array.

car_model_category_encoded = encoder.fit_transform(car_model[category_set]) #this will transform the  categorical into binary/encoded format

```

```python
car_model_category_encoded # we can see we got the numpy array for above operation 
                            #Now we have to tranform following array in dataframe so we can concat it with original dataset
```

```python
car_model_category_encoded = pd.DataFrame(car_model_category_encoded) #here we have transform encoded category into dataframe
```

```python
car_model_category_encoded.head() #from follwoing it seems that we need to bring name of columns rather than index 
                                    #to understand dataset better
```

```python
car_model_category_encoded.columns = encoder.get_feature_names_out(category_set) #this is crucial step to get names 

#we have assigned the columns index new name which we get from encoder.get_features_name
```

```python
encoder.get_feature_names_out(category_set)  #the category seprated
```

```python
car_model_category_encoded #this is the set of dataframe which have categories encoded.
#NOTE -> Cylinder is removed from dataset as no. value of sylinder is only one i.e 4
```

```python
car_model
```

```python
# Now we have to replace car_model original category with car_model_category_encoded and drop the present columns from car_model.

car_model_new = pd.concat([car_model , car_model_category_encoded] , axis=1)
```

```python
car_model_new.columns # we have to drop the categorical columns
```

```python
car_model_new.drop(columns=category_set , inplace=True)
```

milestone point -> Until now our new data set -> car_model_new 

<<--  The categorical columns are encoded succesfully -->>

```python
car_model_new.head()
```

```python
numerical_set = ['age_08_04','km','hp','weight']  #generate set of numerical columns

scaler = StandardScaler()   #  scaling object creation

car_model_new_scale = scaler.fit_transform(car_model_new[numerical_set]) # scale the car_model_new data for numerical variable

```

```python
car_model_new_scale #we got numpy array set
```

```python
# convert above array into data frame
car_model_new_scale = pd.DataFrame(car_model_new_scale)
```

```python
car_model_new_scale.head()

```

```python
#replacing the names of column
car_model_new[numerical_set] = car_model_new_scale
```

##### implentation of machine Learning Model

```python
car_model_new.info()
```

```python
car_model_new.describe()
```

```python
#HeatMap -> 

car_model_new.corr()

plt.figure(figsize=(10,10))

sns.heatmap(car_model_new.corr(),annot=True , annot_kws={"size":7 , "rotation":45} , square=True)
```

```python
from sklearn.model_selection import train_test_split #import of model selection

X = car_model_new.drop(columns=['price']) #X represent data without price | independent paramenter
y = car_model_new['price']
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
#spliting data into training and testing set

X_train , X_test , y_train , y_test = train_test_split(X , y , random_state=42 , test_size=0.2) 

#random_state  = 42 -> this ensure the spliting of the data is same every time we run the code
# test_size = 0.2 -> 20% of data is used for testing and other 80% for training
```

```python
# importing Linear regression model

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Trying to figure out the relationship between y and X so basically we are trainig the model


model.fit(X_train,y_train)
```

```python
#checking the model coefficient and intercept values 
print(f'model Coefficient -> {model.coef_} \n')

print(f'model Intercept -> {model.intercept_}')
```

```python
#let try to predict the value based on above model  result for test data
```

```python
y_prediction = model.predict(X_test)
```

```python
#Let analyse the metric for the prediction and actual result

from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score

mae = mean_absolute_error(y_test , y_prediction)
mse = mean_squared_error(y_test, y_prediction)
r2_score = r2_score(y_test , y_prediction)
```

```python
print(f"mean_absolute_error = {mae} \nmean_squared_error = {mse} \nr2_score = {r2_score}")
```

##### Visualization of findings

```python
import matplotlib.pyplot as plt

plt.scatter(y_prediction , y_test , alpha=0.7 , color='green')
plt.title("Linear Regression Model")
```

```python
#let check manuallly

check = pd.DataFrame({'prediction': y_prediction , 'original' : y_test})
check['diff'] = check['prediction'] - check['original']
check.sort_values('diff')
```

```python
#import necessary libraries
from sklearn.metrics import mean_squared_error , mean_absolute_error ,r2_score
```

```python
mae = mean_absolute_error(y_test , y_prediction)
mse = mean_squared_error(y_test , y_prediction)
R2_score = r2_score(y_test , y_prediction)
root_mse = np.sqrt(mse)

max_car_price = car_model_new["price"].max()
min_car_price = car_model_new["price"].min()

print(f" Range of price : {min_car_price} - {max_car_price}")

print(f' Mean Absolute Error : {mae}\n Mean Squared Error : {mse}\n root_mse : {root_mse}\n R2 Score : {R2_score}')

```

## Lasso and Ridge regression Models

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Lasso_model = Lasso()
```

```python
Lasso_model
```

```python
#Training the model for training set
Lasso_model.fit(X_train,y_train)
```

```python
#predicting the model for test set

y_predict_lasso = Lasso_model.predict(X_test)
```

```python
lasso_graph = sns.scatterplot(x=y_test , y=y_predict_lasso , color="#087830",alpha=.7)
lasso_graph
plt.title("Lasso Regression Model")
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score

#calculate of metrics for lasso
mae_L = mean_absolute_error(y_test , y_predict_lasso)
mse_L = mean_squared_error(y_test , y_predict_lasso)
R2_score_L = r2_score(y_test , y_predict_lasso)
root_mse_L = np.sqrt(mse_L)

max_car_price = car_model_new["price"].max()
min_car_price = car_model_new["price"].min()

print(f" Range of price : {min_car_price} - {max_car_price}")

print(f' Mean Absolute Error Lasso : {mae_L}\n Mean Squared Error Lasso : {mse_L}\n root_mse Lasso : {root_mse_L}\n R2 Score Lasso : {R2_score_L}')



```

Interpretation ->
1. Lasso was reduced the errors as compared to simple linear regression

```python
# Ridge Model

Ridge_model = Ridge()
```

```python
Ridge_model.fit(X_train , y_train)
```

```python
#prediction of value

y_predict_ridge = Ridge_model.predict(X_test)
```

```python
# Analyse errors

#calculate of metrics for Ridge
mae_R = mean_absolute_error(y_test , y_predict_ridge)
mse_R = mean_squared_error(y_test , y_predict_ridge)
R2_score_R = r2_score(y_test , y_predict_ridge)
root_mse_R = np.sqrt(mse_R)

max_car_price = car_model_new["price"].max()
min_car_price = car_model_new["price"].min()

print(f" Range of price : {min_car_price} - {max_car_price}")

print(f' Mean Absolute Error Ridge : {mae_R}\n Mean Squared Error Ridge : {mse_R}\n root_mse Ridge : {root_mse_R}\n R2 Score Ridge : {R2_score_R}')


```

```python
#let compare one by one error from each regression methods
print(f' Mean Absolute Error : {mae}\n Mean Squared Error : {mse}\n root_mse : {root_mse}\n R2 Score : {R2_score} \n')
print(f' Mean Absolute Error Ridge : {mae_R}\n Mean Squared Error Ridge : {mse_R}\n root_mse Ridge : {root_mse_R}\n R2 Score Ridge : {R2_score_R}\n')
print(f' Mean Absolute Error Lasso : {mae_L}\n Mean Squared Error Lasso : {mse_L}\n root_mse Lasso : {root_mse_L}\n R2 Score Lasso : {R2_score_L}')
```
