Keypoints -> 

Tendency ==>
Positive tendency: This might show improving heart rate variability 
or more frequent accelerations, suggesting a healthy or stabilizing condition.
    
Negative tendency: This would include deteriorating signs, such as decreasing variability, increasing decelerations, 
or an abnormal baseline heart rate, potentially indicating worsening fetal health.

NSP ==>

1 (Normal): This indicates a healthy fetal heart rate pattern with no signs of distress. 
The heart rate variability and accelerations are within expected ranges, and no unusual decelerations are observed. 
This status suggests no immediate intervention is required.

2 (Suspect): This status suggests that there are some irregularities in the CTG readings. 
These could include reduced variability or minor decelerations, which may not indicate severe distress 
but require closer monitoring to ensure the fetal condition does not worsen. 
It’s a cautionary status suggesting that the situation should be watched.

3 (Pathological): This classification represents serious concerns with the CTG pattern, potentially indicating fetal distress. 
Examples include repeated or prolonged decelerations, severely low variability, or other patterns that suggest a risk of hypoxia or other complications. 
This value typically calls for urgent evaluation and possibly immediate intervention.

RELATIONSHIP BETWEEN THE COLUMNS

Heatmap are useful to get corrrelation among columns.

before clustering, The data table should be normalized to get an accurate relationship among the variable 

```python
data_normal = (data - data.mean())/data.std()
```

```python
data_normal.head()
```

```python
for column in data_normal.columns:
    plt.hist(data_normal[column])
    plt.show()
```

```python
data_normal_matrix = data_normal.corr()
```

```python
data_normal_matrix.head()
```

```python
sns.heatmap(data_normal_matrix , cmap='coolwarm')
```

```python
sns.clustermap(data_normal_matrix , cmap='RdBu' ,center=0 , annot=True)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data_normal_matrix , annot=True , fmt='.1f' , annot_kws={"size":8} , cmap='coolwarm')
plt.show()
```

conclusion ->>
