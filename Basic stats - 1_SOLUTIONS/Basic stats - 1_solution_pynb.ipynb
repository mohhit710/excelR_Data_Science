#### Standardization of Volume 

```python
vol = sales["volume"]
vol_mean = vol.mean()
vol_sd = vol.std()
print(vol_mean , vol_sd)

vol_stdz = (sales["volume"] - vol_mean)/vol_sd

print(vol_stdz)

vol_stdz.std()
```

```python
fig,ax = plt.subplots(1,2 , figsize = (10,5))

fig.suptitle("Volume")
#fig.patch.set_facecolor("#57dcbe")
fig.patch.set_edgecolor("black")
fig.patch.set_linewidth(3)

ax[0].hist(vol , color="grey")
ax[0].set_title("Original Dataset")


ax[1].hist(vol_stdz , color="black")
ax[1].set_title("Standardised Dataset")

plt.show()
```

#### Standardization of avg_price

```python
avg_price = sales["avg_price"]
avg_price_mean = avg_price.mean()
avg_price_sd = avg_price.std()
print(avg_price_mean , avg_price_sd)

avg_price_stdz = (avg_price - avg_price_mean)/avg_price_sd

print(avg_price_stdz)
```

```python
fig,ax = plt.subplots(1,2 , figsize = (10,5))

fig.suptitle("avg_price")
#fig.patch.set_facecolor("#57dcbe")
fig.patch.set_edgecolor("black")
fig.patch.set_linewidth(3)

ax[0].hist(avg_price , color="grey")
ax[0].set_title("Original Dataset")


ax[1].hist(avg_price_stdz , color="black")
ax[1].set_title("Standardised Dataset")

plt.show()
```

#### Standardization of total_sales_value

```python
fig,ax = plt.subplots(1,2 , figsize = (10,5))

fig.suptitle("total_sales_value")
#fig.patch.set_facecolor("#57dcbe")
fig.patch.set_edgecolor("black")
fig.patch.set_linewidth(3)

ax[0].hist(total_sales_value , color="grey")
ax[0].set_title("Original Dataset")


ax[1].hist(total_sales_value_stdz , color="black")
ax[1].set_title("Standardised Dataset")

plt.show()
```
